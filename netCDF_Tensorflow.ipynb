  1.3 - Analysis of Dataset Structure: Dimensions, Coordinates, and Variables in a NetCDF Climate File:
  
      1.3.1 - The dimensions in the dataset are as follows:

        * 'time_counter': This dimension represents the time component of the data. Given that its length is 12, it likely represents monthly data for one year.

        * 'bnds': Typically, 'bnds' dimension is associated with boundary variables that describe the boundaries of cell intervals on grid dimensions. The purpose of 'bnds' is to define the bounds of grid cells. However, it's not used by any of the data variables in your output.

        * 'lon': This dimension represents the longitude component. The length of 384 suggests that your longitude is sampled at 384 evenly spaced points.

        * 'lat': This dimension represents the latitude component. The length of 221 suggests that your latitude is sampled at 221 evenly spaced points.

      1.3.2 -The coordinates in the dataset are:

        * 'time_counter': These are the datetime64[ns] timestamps that correspond to each time step in the file.

        * 'lon': A series of float32 values representing evenly spaced intervals of geographic longitude.

        * 'lat': A series of float32 values representing evenly spaced intervals of geographic latitude.

      1.3.3 - The primary variables in the dataset are:

        * 'tsol': A 3D variable that varies with (time_counter, lat, lon). It may represent solar temperature.

        * 't2m': A 3D variable that varies with (time_counter, lat, lon). This represents the near-surface air temperature, specifically the temperature 2 meters above the ground.

        * 'psol': A 3D variable that varies with (time_counter, lat, lon). This could represent solar pressure.

```python
import matplotlib.pyplot as plt
```

```python
# 2.1 Average Precipitation for 1980

# Load the 1980 dataset
data_1980 = xr.open_dataset('sim-Morocco-1980.nc')

# Calculate the average precipitation for 1980
precip_1980 = data_1980['precip'].mean(dim='time_counter')

# Visualize the average precipitation for 1980
precip_1980.plot(cmap='jet')
plt.title('Average Precipitation for 1980')
plt.show()
```

```python
# 2.2 Average Precipitation for 2014

# Load the 2014 dataset
data_2014 = xr.open_dataset('sim-Morocco-2014.nc')

# Calculate the average precipitation for 2014
precip_2014 = data_2014['precip'].mean(dim='time_counter')

# Visualize the average precipitation for 2014
precip_2014.plot(cmap='jet')
plt.title('Average Precipitation for 2014')
plt.show()
```

```python
# 2.3 Comparison of Precipitation Over Morocco for 1980 and 2014

# Specify the latitude and longitude range for each data 1980 and 2014 (tehy are the same)
lat_range_1980 = data_1980['lat']
lon_range_1980 = data_1980['lon']

lat_range_2014 = data_2014['lat']
lon_range_2014 = data_2014['lon']

# Extract the data
morocco_precip_1980 = precip_1980.sel(lat=lat_range_1980, lon=lon_range_1980)
morocco_precip_2014 = precip_2014.sel(lat=lat_range_2014, lon=lon_range_2014)


# Visualize precipitation over Morocco for 1980 and 2014
fig, axes = plt.subplots(nrows=2, figsize=(12, 8))

morocco_precip_1980.plot(ax=axes[0], cmap='jet')
axes[0].set_title('Average Precipitation over Morocco for 1980')

morocco_precip_2014.plot(ax=axes[1], cmap='jet')
axes[1].set_title('Average Precipitation over Morocco for 2014')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import zoom
from matplotlib.ticker import FuncFormatter
from skimage.metrics import peak_signal_noise_ratio
import tensorflow as tf

# Function to format tick labels as integers
def format_ticks(value, _):
    return f'{int(value):,}'

# Function to normalize image data to the valid range [0, 1]
def normalize_image(image):
    image_min = np.min(image)
    image_max = np.max(image)
    return (image - image_min) / (image_max - image_min)

fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Plot Original Image
axs[0].imshow(normalize_image(tf.squeeze(lr_image)))
axs[0].set_title('Original Image')
axs[0].set_xlabel('X-axis')
axs[0].set_ylabel('Y-axis')
axs[0].get_xaxis().set_major_formatter(FuncFormatter(format_ticks))
axs[0].get_yaxis().set_major_formatter(FuncFormatter(format_ticks))

# Plot Predicted Image
axs[1].imshow(normalize_image(tf.squeeze(hrdl_image)))
axs[1].set_title('Predicted Image')
axs[1].set_xlabel('X-axis')
axs[1].set_ylabel('Y-axis')
axs[1].get_xaxis().set_major_formatter(FuncFormatter(format_ticks))
axs[1].get_yaxis().set_major_formatter(FuncFormatter(format_ticks))

# Resize the predicted image to match the shape of the original image
predicted_image_resized = zoom(hrdl_image, (original_image.shape[0] / hrdl_image.shape[0], original_image.shape[1] / hrdl_image.shape[1], 1))
predicted_image_resized_np = np.asarray(predicted_image_resized)

# Add a comparison result based on the compare_images function
psnr_value = peak_signal_noise_ratio(original_image, predicted_image_resized_np)
axs[1].text(0.5, -0.15, f'PSNR: {psnr_value:.2f}', ha='center', transform=axs[1].transAxes)

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Show the plots
plt.show()
```
